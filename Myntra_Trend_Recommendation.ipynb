# **Loading the Dataset**

*2) User_data consisting of user details like purchase history and preferences of the user.*

```python
from IPython.display import display, HTML

import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics.pairwise import cosine_similarity

product = pd.read_csv('/content/product_data.csv')
user = pd.read_csv('/content/user_data.csv')
```

# **Data Preprocessing**

```python
product.dropna(inplace=True)
user.dropna(inplace=True)
```

Maping product id with product name

```python
product_to_index = {product_name: index for index, product_name in enumerate(product['name'].unique())}
```

# **Collaborative Filtering Method** - *recommendation based on purchase history and browsing behavior*

```python

def collaborative_filtering(user_id, user_data, product_data):
    user_data_subset = user_data[user_data['user_id'] == user_id]
    if user_data_subset.empty:
        return []
    user_data_row = user_data_subset.iloc[0]

    interactions = user_data_row[['purchase_history.1', 'purchase_history.2', 'purchase_history.3',
                                  'browsing_behavior.1', 'browsing_behavior.2', 'browsing_behavior.3']].values.flatten()

    interactions_indices = [product_to_index.get(product, -1) for product in interactions if product in product_to_index]

    similar_users = user_data[user_data['user_id'] != user_id]
    similarities = []

    for _, row in similar_users.iterrows():
        other_interactions = row[['purchase_history.1', 'purchase_history.2', 'purchase_history.3',
                                  'browsing_behavior.1', 'browsing_behavior.2', 'browsing_behavior.3']].values.flatten()

        other_interactions_indices = [product_to_index.get(product, -1) for product in other_interactions if product in product_to_index]

        min_length = min(len(interactions_indices), len(other_interactions_indices))
        if min_length > 0:
            similarity = cosine_similarity([interactions_indices[:min_length]], [other_interactions_indices[:min_length]])[0][0]
            similarities.append((row['user_id'], similarity))

    similarities.sort(key=lambda x: x[1], reverse=True)
    top_similar_users = similarities[:5]

    recommended_products = []
    for user_id, _ in top_similar_users:
        user_interactions = user_data[user_data['user_id'] == user_id][['purchase_history.1', 'purchase_history.2', 'purchase_history.3',
                                                                        'browsing_behavior.1', 'browsing_behavior.2', 'browsing_behavior.3']].values.flatten()
        for product in user_interactions:
            if product not in interactions:
                recommended_products.append(product)

    return recommended_products
```

# **Content-Based Filtering** - *Recomndation based on user preferences and product attributes*

```python
def content_based_filtering(user_id, product_data):
    user_preferences = user[user['user_id'] == user_id].iloc[0][['preferences.1', 'preferences.2']]

    filtered_products = product_data[(product_data['p_attributes'].str.contains(user_preferences['preferences.1'])) &
                                     (product_data['avg_rating'] >= 4)]


    filtered_products = filtered_products[['p_id', 'img', 'avg_rating', 'ratingCount', 'name']].head(9)
    return filtered_products[['p_id', 'img', 'name']]
```

# **Hybrid Recommendation System** - *combining Collaborative and Content-Based Filtering*

```python
def hybrid_recommendation(user_id, user_data, product_data):
    collaborative_recommended = collaborative_filtering(user_id, user_data, product_data)
    content_based_recommended = content_based_filtering(user_id, product_data)['p_id'].tolist()

    hybrid_recommended = list(set(collaborative_recommended + content_based_recommended))

    return hybrid_recommended
```

# **Function to evaluate Recommendation system**

```python
def evaluate_recommendations(user_data, product_data):
    evaluation_results = []

    for user_id in user_data['user_id'].unique():
        true_products = user_data[user_data['user_id'] == user_id][['true_product.1', 'true_product.2', 'true_product.3']].values.flatten()
        true_products = [product for product in true_products if pd.notna(product)]  # Remove NaN values


        recommended_products_collaborative = collaborative_filtering(user_id, user_data, product_data)

        recommended_products_content = content_based_filtering(user_id, product_data)['p_id'].tolist()

        recommended_products_hybrid = hybrid_recommendation(user_id, user_data, product_data)

        def get_binary_relevance(products, reference_products):
            return [1 if product in reference_products else 0 for product in products]

        binary_true_products = get_binary_relevance(product_data['name'].unique(), true_products)
        binary_recommended_collaborative = get_binary_relevance(product_data['name'].unique(), recommended_products_collaborative)
        binary_recommended_content = get_binary_relevance(product_data['name'].unique(), recommended_products_content)
        binary_recommended_hybrid = get_binary_relevance(product_data['name'].unique(), recommended_products_hybrid)

        # Calculate accuracy metrics only
        accuracy_collab = accuracy_score(binary_true_products, binary_recommended_collaborative)
        accuracy_content = accuracy_score(binary_true_products, binary_recommended_content)
        accuracy_hybrid = accuracy_score(binary_true_products, binary_recommended_hybrid)

        evaluation_results.append({
            'User ID': user_id,
            'True Products': true_products,
            'Collaborative Recommended Products': recommended_products_collaborative,
            'Content-Based Recommended Products': recommended_products_content,
            'Hybrid Recommended Products': recommended_products_hybrid,
            'Collaborative Filtering - Accuracy': accuracy_collab,
            'Content-Based Filtering - Accuracy': accuracy_content,
            'Hybrid Filtering - Accuracy': accuracy_hybrid,
        })

    return evaluation_results
```

*Accuracy of the recomendation system*

```python
#Accuracy

evaluation_results = evaluate_recommendations(user, product)

total_accuracy_collab = sum(result['Collaborative Filtering - Accuracy'] for result in evaluation_results) / len(evaluation_results)
total_accuracy_content = sum(result['Content-Based Filtering - Accuracy'] for result in evaluation_results) / len(evaluation_results)
total_accuracy_hybrid = sum(result['Hybrid Filtering - Accuracy'] for result in evaluation_results) / len(evaluation_results)

print(f"Overall Collaborative Filtering Accuracy: {total_accuracy_collab}")
print(f"Overall Content-Based Filtering Accuracy: {total_accuracy_content}")
print(f"Overall Hybrid Filtering Accuracy: {total_accuracy_hybrid}")
```
